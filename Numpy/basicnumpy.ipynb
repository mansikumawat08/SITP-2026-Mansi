## NumPy in Python

NumPy (Numerical Python) is a powerful Python library used for numerical and mathematical operations. It provides support for arrays, matrices, and many built-in functions for fast calculations. NumPy is faster and more memory-efficient than normal Python lists.

### Features

* Supports N-dimensional arrays
* Performs mathematical operations easily
* Useful for matrix and statistical calculations
* Faster data processing

### Uses

NumPy is widely used in Data Science, Machine Learning, and Data Analytics. It also works with libraries like Pandas and Matplotlib.

```python
# ==============================
# NumPy Array Operations Example
# ==============================

import numpy as np

arr1 = np.array([10, 20, 30, 40])
arr2 = np.array([2, 4, 6, 8])

print("Array 1 :", arr1)
print("Array 2 :", arr2)

print("\n========== OPERATIONS ==========\n")

# Multiplication
print("1️⃣ Multiplication (*)")
print("arr1 * arr2 =", arr1 * arr2)

# Division
print("\n2️⃣ Division (/)")
print("arr1 / arr2 =", arr1 / arr2)

# Addition
print("\n3️⃣ Addition (+)")
print("arr1 + arr2 =", arr1 + arr2)

# Subtraction
print("\n4️⃣ Subtraction (-)")
print("arr1 - arr2 =", arr1 - arr2)

# Positive Sign with Multiplication
print("\n5️⃣ Multiplication with Unary Plus (* +)")
print("arr1 * +arr2 =", arr1 * +arr2)

# Floor Division
print("\n6️⃣ Floor Division (//)")
print("arr1 // +arr2 =", arr1 // +arr2)

print("\n✅ All operations completed successfully!")
```
