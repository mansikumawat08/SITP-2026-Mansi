```python
# 1 class number series using loop
class NumberSeries:
    def print_number(self):
     for x in range(1,101):
        print(x)
        
obj =NumberSeries()
obj.print_number()
```

```python
#2 class check even odd 
class EvenOdd:
    def print_number(self):
        for i in range(1,51):
            if i%2 == 0:
                print(i,"even")
            else:
                print(i,"odd")
obj=EvenOdd()               
obj.print_number()
```
