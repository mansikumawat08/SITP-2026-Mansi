### Class in Python

A class is a blueprint or template used to create objects.

It defines:

Data/Properties → variables
Behavior/Functions → methods

Think of a class like a design of a car.
The actual cars made from that design are objects.

## Syntax

    class Student:
        pass

class Student:
    name = "Ritika"
    branch = "AIML"

# Here:
# Student → class name
# name and branch → attributes (variables inside class)

| Pillar            | Definition                                                                                          | How Achieved                                                                        | Main Purpose                            | Main Keywords / Concepts                             | Example                               |
| ----------------- | --------------------------------------------------------------------------------------------------- | ----------------------------------------------------------------------------------- | --------------------------------------- | ---------------------------------------------------- | ------------------------------------- |
| **Encapsulation** | Wrapping data and methods together into a single unit (class) and restricting direct access to data | By using classes and access modifiers like public, protected, and private variables | Data security and organized code        | Class, Object, `self`, private variable (`__var`)    | `self.__marks = 90`                   |
| **Inheritance**   | One class acquires properties and methods of another class                                          | By creating a child class from a parent class                                       | Code reusability and reduced redundancy | Parent class, Child class, `super()`                 | `class Dog(Animal)`                   |
| **Polymorphism**  | Same method/operator behaves differently in different situations                                    | By method overriding, method overloading, and operator overloading                  | Flexibility and cleaner code            | Method overriding, overloading, operator overloading | `sound()` method in different classes |
| **Abstraction**   | Hiding internal implementation details and showing only essential features                          | By using abstract classes and abstract methods                                      | Reduce complexity and improve security  | `ABC`, `@abstractmethod`                             | Abstract `Shape` class                |
