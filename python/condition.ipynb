```python
#Program to check whether a number id positive , negative or zero

num = float(input("Enter a number: "))

if num > 0:
    print("the number is positive ")
elif num < 0: 
    print("The number is negative")
else:
    print("The number is zero")
    
```

```python
fruits = ["apple","banana","cherry"]
for x in fruits:
    if x == "banana":
        continue# skips the element passed
    print(x)
```
