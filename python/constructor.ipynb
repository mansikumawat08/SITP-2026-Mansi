```python
class Student:
    def __init__(self,name,branch,roll_number,classes):
        self.name=name
        self.branch=branch
        self.roll_number=roll_number
        self.classes=classes

    def display(self):
        print(self.name)

clg=Student(name="Mansi",branch="IT",roll_number=18,classes="3rd yr")
clg.display()
```
