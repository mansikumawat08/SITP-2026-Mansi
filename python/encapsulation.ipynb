ENCAPSULATION

Encapsulation in Python

Encapsulation means wrapping data (variables) and methods (functions) together inside a class and restricting direct access to some data for security and better control.

Simple Definition

“Encapsulation is the process of hiding data inside a class and allowing access through methods.”

```python
class College:
    college_name = "GWECA"### public by default
    college_garden = "yes"
    college_account = 123478890

clg = College()

clg.college_account()
clg.college_garden()
clg.college_name()
```

```python
class College:
    college_name = "GWECA"### public by default
    college_garden = "yes"
    ___college_account = 123478890## private (using underscore)

clg = College()

clg.___college_account
```

```python
class College:
    college_name = "GWECA"### public by default
    college_garden = "yes"
    ___college_account = 123478890## private (using underscore)
    
    def display(self):
        return self.___college_account## way to acces private key using method
    

clg = College()
clg.display()
```

```python
class College:
    college_name = "GWECA"### public by default
    college_garden = "yes"
    ___college_account = 123478890## private (using underscore)
    
    def display(self,a,b):
        print(a+b)
        return self.___college_account
    

clg = College()
clg.display(a=10,b=20)    
```
