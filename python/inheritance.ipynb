* Inheritance is a feature of OOP (Object-Oriented Programming) where one class can use the properties and methods of another class.

* Parent Class / Base Class → The class whose properties are inherited.
* Child Class / Derived Class → The class that inherits properties.

        Inheritance helps in:

* Code reusability
* Reducing duplicate code
* Easy maintenance


Types of Inheritance in Python


| Type                     | Meaning                                    |
| ------------------------ | ------------------------------------------ |
| Single Inheritance       | One parent → One child                     |
| Multiple Inheritance     | Multiple parents → One child               |
| Multilevel Inheritance   | Grandparent → Parent → Child               |
| Hierarchical Inheritance | One parent → Multiple children             |
| Hybrid Inheritance       | Combination of different inheritance types |

Single Inheritance

```python
# single inheritance

class College:
    pass
class Student(College):
    pass
```

```python
class College:
    college_name = "GWECA"
    college_address ="Ajmer"
    college_sp =42
    def show(self):
        print(self.college_name)
        print(self.college_address)
        print(self.college_sp)

class Student(College):
    student_name ="Mansi"
    student_rollnumber=123789
    student_number=1234789012
    student_branch ="IT"
st =Student()
st.show()
print(st.student_branch)
```

Multilevel Inheritance

```python
#Multilevel inheritance
class Person:
    pass
class Teacher(Person):
    pass
class HOD(Teacher):
    pass
```

```python
#Multilevel inheritance
class Person:
    def person_info(self):
        print("I am a person")

class Teacher(Person):
    def teacher_info(self):
        print("I am a teacher")

class HOD(Teacher):
    def hod_info(self):
        print("i am HOD")

h = HOD()
print(h.hod_info())
print(h.person_info())
print(h.teacher_info())
```

Multiple Inheritance

```python
#multiple inheritance
class Developer:
    def coding(self):
        print("ye code likhega")

class Manager:
    def meeting(self):
        print("meeting leta h")

class Teamleader(Developer,Manager):
    def project(self):
        print("project ko handle krta h")

team =Teamleader()

team.coding()

```

```python
team.meeting()
```

```python
team.project()
```

```python
class Employee1:
    def sales(self):
        print("works on sales data")

class Employee2:
    def customer(self):
        print("works on customer data")

class DataManager(Employee1,Employee2):
    def analytics(self):
        print("analyses the data")

analyse= DataManager()
analyse.sales()
```

```python
analyse.customer()
```

```python
analyse.analytics()
```

```python
class Developer:
    def coding(self):
        print("ye code likhega")

class Manager(Developer):
    def meeting(self):
        print("meeting leta h")

class Teamleader(Developer):
    def project(self):
        print("project ko handle krta h")

team =Teamleader()
man = Manager()


team.coding()
man.coding()
# yaha p 2 child class ek parent class ko access krti h
```

HYBRID INHERITANCE

```python
class Developer:
    def coding(self):
        print("ye code likhega")

class Manager(Developer):
    def meeting(self):
        print("meeting leta h")

class Teamleader(Developer):
    def project(self):
        print("project ko handle krta h")

class Designer(Manager,Teamleader):
    def design(self):
        print("this person designs")

team = Teamleader()
man = Manager()
des = Designer()

team.coding()
man.coding()
des.project()
des.meeting()
```
